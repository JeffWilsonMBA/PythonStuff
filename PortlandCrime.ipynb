```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Reference the file where the CSV is located
crime_csv_path = os.path.join ("..", "Resources", "crime_incident_data2017.csv")

# Import the data into a Pandas DataFrame
crime_df = pd.read_csv(crime_csv_path) # read_excel, etc.
crime_df.head()
```

```python
# Get the count of rows within the datafrom in order to determine if there are any null values
crime_df.count()
```

```python
# Drop the rows that contain null values
no_null = crime_df.dropna(how='any')
no_null.count()
# () means it's taking an action, verb
# No () it's an attribute or adjective
# crime_df.copy - this creates a copy of the dataFrame
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
no_null["Offense Type"].value_counts()
```

```python
# Combining similar offenses together
no_null["Offense Type"] = no_null["Offense Type"].replace({"Commercial Sex Acts":"Prostitution","Assisting or Promoting Prostitution":"Prostitution"})
no_null["Offense Type"].value_counts()
```

```python
# Create a new DataFrame that looks into a specific neighborhood
vernon_crime_df = no_null.loc[no_null["Neighborhood"] == "Vernon"]
vernon_crime_df
```
