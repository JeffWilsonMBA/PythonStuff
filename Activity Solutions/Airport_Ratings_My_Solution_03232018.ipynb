```python
# Dependencies
import os
import pandas as pd
import numpy as np
import requests
import json

# Import API key
from config import api_key
```

```python
# Import cities file into the cities_pd DataFrame
city = os.path.join("cities.csv")
cities = pd.read_csv(city)
cities_pd = pd.DataFrame(cities)
cities_pd.head()
```

```python
# Add columns for lat, lng, airport name, airport address, airport rating.
# Hint: Be sure to specify "" as the initial value for each column.
cities_pd["Lat"] = ""
cities_pd["Lng"] = ""
cities_pd["Airport Name"] = ""
cities_pd["Airport Address"] = ""
cities_pd["Airport Rating"] = ""
cities_pd.head()
```

```python
params = {"key": api_key}
# Loop through the cities_pd and get the lat/long for each city
# Hint: `for index, row in cities_pd.iterrows():`
for index, row in cities_pd.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    city = row['City']
    state = row['State']
    
    # Update address key value
    params['address'] = f"{city},{state}"
    
    # Make request and print urls
    # Hint: `requests.get(target_url).json()`
    cities_lat_lng = requests.get(base_url,params=params)
    print(cities_lat_lng.url)
    
    # Convert to json
    cities_lat_lng = cities_lat_lng.json()
    
    # Hint: `cities_pd.set_value()`
    cities_pd.set_value(
        index, "Lat", cities_lat_lng["results"][0]["geometry"]["location"]["lat"])
    cities_pd.set_value(
        index, "Lng", cities_lat_lng["results"][0]["geometry"]["location"]["lng"])
    
# print(json.dumps(cities_lat_lng, indent = 4, sort_keys = True))

cities_pd.head()
```

```python
# Use the lat/lng we recovered to identify airports
# Hint: Use the keyword search for "international+airport"
params = {
    "radius": 50000,
    "types": "airport",
    "keyword": "international airport",
    "key": api_key
}

# Hint: Code is similar to above block (i.e. you will need to use cities_pd.iterrows() again)
for index, row in cities_pd.iterrows():
    # Get lat and lng for df
    lat = row["Lat"]
    lng = row["Lng"]
    
    params["location"] = f"{lat},{lng}"
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    name_address = requests.get(base_url,params = params)
    print(name_address.url)
    
    name_address = name_address.json()
    # print(json.dumps(name_address, indent=4, sort_keys=True))
    try:
        cities_pd.set_value(index, "Airport Name",
                            name_address["results"][0]["name"])
        cities_pd.set_value(index, "Airport Address",
                            name_address["results"][0]["vicinity"])
        cities_pd.set_value(index, "Airport Rating",
                            name_address["results"][0]["rating"])
    except (KeyError, IndexError):
        print("Missing field/result...skipping")
                    
# Hint: Use a try-except to handle airports that are missing ratings




```

```python
# Save Data to csv
cities_pd.to_csv("Airport_Data.csv")

# Visualize to confirm airport data appears
cities_pd.head()
```
