```python
# Dependencies
import json
import requests

# Load JSON
url = "http://nyt-mongo-scraper.herokuapp.com/api/headlines"
data = requests.get(url)
print(data.status_code)
```

```python
data_json = data.json()
data_json
```

```python
print(f"The first response is {json.dumps(data_json[0],indent=2)}.")
print(f"The last response is {json.dumps(data_json[-1], indent=2)}.")
```

```python
print(f"We received {len(data_json)} responses.")
```
